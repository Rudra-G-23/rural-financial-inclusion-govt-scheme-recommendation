# Possible Reason behind the null value 
- During `Str -> Int` casting time

```python
import polars as pl

df = pl.DataFrame({
    "a": [
        "1", 
        "2", 
        "x",      #  ← not a number
        "3",
        "",       #  ← empty string
        " 4",     #  ← leading space!
        " "
    ]
})

display(df)
```

```python

df2 = df.with_columns(
    pl.col("a").cast(pl.Int64, strict=False)
)

display(df2)
```

```python
# We notice that only null value are non integer
```
