# Functions

```python
def draw_pie(col: str) -> None:
    df = pdf[col].value_counts().to_pandas()
    display(df)

    plt.pie(
        df['count'],
        labels=df[col],
        autopct='%1.2f%%'
    )
    plt.title(f"{col} Percentage")
    plt.show()
    print({'=='*45})
```
